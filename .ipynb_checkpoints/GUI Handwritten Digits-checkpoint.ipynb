```python
import tkinter as tk
from PIL import ImageTk,Image,ImageDraw
import numpy as np
import cv2 as cv
import joblib

win=tk.Tk()#main window

w,h=500,500
canvas=tk.Canvas(win,width=w,height=h,bg='white')
canvas.grid(row=0,column=0,columnspan=4)
count=0

model=joblib.load('KNN - Handwritten Digits.sav')

def save():
    global count
    imgArray=np.array(img)
    imgArray=cv.resize(imgArray,(8,8))
    cv.imwrite('data/'+str(count)+'.jpg',imgArray)#saving the image
    count=count+1
    
buttonSave=tk.Button(win,text='SAVE',bg='yellow', fg='white',font='Helvetica 20 bold',command=save)
buttonSave.grid(row=1,column=0)

def predict():
    #preprocess the image
    imgArray=np.array(img)
    #RGB to grey scale
    imgArray=cv.cvtColor(imgArray,cv.COLOR_RGB2GRAY)
    #resize
    imgArray=cv.resize(imgArray,(8,8))
    #flatten the image
    imgArray=np.reshape(imgArray,(1,64))
    #change pixel value to 4bit number
    imgArray=(imgArray/255)*15
    result=model.predict(imgArray)
    #update label
    labelStatus.config(text="PREDICTED DIGIT:"+str(result))
    
buttonPredict=tk.Button(win,text='PREDICT',bg='green', fg='white',font='Helvetica 20 bold',command=predict)
buttonPredict.grid(row=1,column=1)

def clear():
    global img,imageDraw
    canvas.delete('all')
    img=Image.new('RGB',(w,h),(0,0,0))
    imageDraw=ImageDraw.Draw(img)
    
buttonClear=tk.Button(win,text='CLEAR',bg='blue', fg='white',font='Helvetica 20 bold',command=clear)
buttonClear.grid(row=1,column=2)

buttonExit=tk.Button(win,text='EXIT',bg='red', fg='white',font='Helvetica 20 bold',command=win.destroy)
buttonExit.grid(row=1,column=3)

labelStatus=tk.Label(win,text="PREDICTED DIGIT: NONE", bg='white', fg='black', font='Helvetica 24 bold')
labelStatus.grid(row=2,column=0,columnspan=4)
    
img=Image.new('RGB',(w,h),(0,0,0))
imageDraw=ImageDraw.Draw(img)

def eventFunction(event):
    x = event.x
    y = event.y
    x1 = x - 20
    y1 = y - 20
    x2 = x + 20
    y2 = y + 20
    canvas.create_oval((x1, y1, x2, y2), fill = 'black')
    imageDraw.ellipse((x1, y1, x2, y2), fill = 'white')
    
canvas.bind('<B1-Motion>',eventFunction)

win.mainloop()
```
